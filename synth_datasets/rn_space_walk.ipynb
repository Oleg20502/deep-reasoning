```python
import numpy as np

def vec_to_int64_vec(vec):
    return np.array(vec, dtype=np.int64)

def vec_to_str(vec):
    return str(tuple(int(i) for i in vec))


def generate_example(number_of_walks = 10000, dims=10, mean=None, cov=None):
    """Problem moves: (+1, -2, +3); (); ().
    Answer: (1, 0, -1)
    """
    if mean is None:
        mean = np.zeros(dims)
    if cov is None:
        cov = np.eye(dims) * 4
    pos = np.zeros(dims)
    problem_str = ""
    for i in range(number_of_walks):
        step_vec = vec_to_int64_vec(np.random.multivariate_normal(mean, cov))
        problem_str += vec_to_str(step_vec) + '; '
        pos += step_vec
    return problem_str, vec_to_str(pos)
```

```python
for i in range(10):
    number_of_walks = 2 + i // 10 
    dims = 5 * (i % 10) + 5
    task, answer = generate_example(number_of_walks, dims)
    print(task)
    print(answer)
```
